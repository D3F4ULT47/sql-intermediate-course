```python
%%sql
SELECT
p.categoryname,
SUM(CASE WHEN s.orderdate BETWEEN '2022-01-01' AND '2022-12-31' THEN s.quantity * s.netprice * s.exchangerate ELSE 0 END) AS net_revenue
FROM
 sales s
LEFT JOIN product p ON p.productkey = s.productkey
GROUP BY
p.categoryname

```
